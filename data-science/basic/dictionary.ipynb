# Matrix

Let $A$ be a 2-dimensional matrix,

$$A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$$

$$\mathrm{trace(A)} = a + d$$
$$\mathrm{determinant(A)} |A| := ad - bc$$
