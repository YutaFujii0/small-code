## polinominal curve fitting

* simple regression problem (thus, supervised learning!)
* curve fitting to predict $\hat{t}$ given $\hat{x}$
* $\hat{t}$ from given $\hat{x}$ has uncertainty
* fitting is achieved by minimizing **error function**
* flamework to evaluate that uncertainty(1.2)
* a dicision theory enable us to predict under appropriate rule, using probability framework

**nomal equation or gradient decent**

* feature scaling doesn't matter for nomal equation
* without feature scaling, gradient decent may have huge steps until convergence

$$x=\frac{x-\mu}{\sigma}$$ 

**linear model**
curve fitting: $t=y(x, \mathbf{w})$

$$y(x,\mathbf{w}) = w_0 + w_1x + w_2x^2 + \dots + w_Mx^M = \sum_{j=0}^{M}w_jx^j$$

* non-linear function for $x$
* **important!** linear function for $\mathbf{w}$
* 

**error function**
sum-of-squeres error:

$$E(\mathbf{w}) = \frac{1}{2}\sum_{n=1}^{M}\{y(x_n,\mathbf{w}) - t_n\}^2$$

* function that has order of $\mathbf{w}$ squere
* thus, it has a unique solution $\mathbf{w*}$ to minimize its value

$$\mathbf{w*} = TODO$$

**issue-Model comparison/Model selection**

* The larger M is, the better $\mathbf{w}$ we can get? NO!!!
* when M equals to the num of data, error function can be 0 at $\mathbf{w} = \mathbf{w*}$, but such $y(x, \mathbf{w})$ looks far from that should be
* this is called **over fitting**

To avoid over fitting

1. roughly speaking, data > 5 * M then over fitting won't be an issue
1. regularization: give a penalty to error function
    * shrinkage (statistics)
    * weight dacay (neural network)
    * ridge regression ($\mathbf{w}$ squered penalty)
1. Bayesian approach(3.4)
