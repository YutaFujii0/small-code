```python
import quandl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
bigmac_us = quandl.get("ECONOMIST/BIGMAC_USA", returns="numpy")
```

```python
sentiment = quandl.get("AAII/AAII_SENTIMENT", returns="pandas")
sentiment.head(5)
```

```python
# x = sentiment["Date"]
y_bull = sentiment["Bullish"]
y_bear = sentiment["Bearish"]
y_bear.head(5)
```

```python
x = sentiment.index
type(x)
```

```python
fig, ax = plt.subplots()
line1, = ax.plot(x, y_bull, '--', linewidth=2, label='Bullish')
line2, = ax.plot(x, y_bear, '-', linewidth=2, label='Bearish')
ax.legend(loc='lower right')
plt.show()
```

```python
y_diff = y_bull - y_bear
```

```python
fig2, ax2 = plt.subplots()
line, = ax2.plot(x, y_diff, '-', linewidth=1, label="sentiment")
ax2.legend(loc="upper left")
plt.show()
```

```python
begin_date = datetime(2012,1,1, 0,0)
x_new = x[(x > begin_date)]
data_new = sentiment[(sentiment.index > begin_date)]
data_new.head(5)
```

```python
y_diff_new = data_new["Bullish"] - data_new["Bearish"]
```

```python
fig3, ax3 = plt.subplots()
line3, = ax3.plot(x_new, y_diff_new, '-', linewidth=1, label="sentiment")
ax3.legend(loc="upper left")
plt.show()
```

```python
a = np.arange(0,1,0.01)
b1 = np.sin(a * 2 * np.pi)
b2 = np.cos(a * 2 * np.pi)
plt.plot(a, b1, label="sin(x)")
plt.plot(a, b2, label="cos(x)")
plt.legend(loc="lower right")
plt.show()
```
